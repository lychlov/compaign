```python
import pandas as pd
import numpy
```

```python
funds = pd.read_csv(r'F:\downloads\hypernum_fund.csv')
projects = pd.read_csv(r'F:\downloads\hypernum_project.csv')
```

```python
projects
```

```python
funds_set = set()
funds_dict = dict()
count = 0
for index, row in projects.iterrows():
    for fund in row['funds'].replace("[", '').replace("]", '').replace("'", '').split(','):
        count += 1
        funds_dict[fund] = funds_dict[fund] + 1 if funds_dict.get(fund,None) else 1
        funds_set.add(fund)
```
