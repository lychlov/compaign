```python
all_alarm['a_flag']=all_alarm['CLEAR_USER'].apply(lambda x:str(x)=='AUTOMATIC')
```

```python
all_alarm.head()
```

```python
all_alarm['flag']=all_alarm['SOURCE_ID'].apply(lambda x:str(x).startswith('MIX-'))
all_alarm.head()

```

```python
filtered = all_alarm[(all_alarm.flag==False)&(all_alarm.a_flag==True)]
filtered.head()
```

```python
filtered.info()
```

```python
filtered['START_TIME_PD'] =pd.to_datetime(filtered['START_TIME'])
filtered['SAVE_TIME_PD'] =pd.to_datetime(filtered['SAVE_TIME'])
```

```python
filtered.head()
```

```python
import time
filtered['START_TIME_timestamp'] = filtered['START_TIME'].apply(lambda x:time.mktime(time.strptime(x,'%Y-%m-%d %H:%M:%S')))
filtered['SAVE_TIME_timestamp'] = filtered['SAVE_TIME'].apply(lambda x:time.mktime(time.strptime(x,'%Y-%m-%d %H:%M:%S')))
```

```python
filtered.eval('wucha = SAVE_TIME_timestamp - START_TIME_timestamp' , inplace=True)
filtered.head()
```

```python
filtered[(filtered.wucha>=600)].count()['wucha']
```

```python
filtered.head()
```

```python
filtered['wucha'].max()
```

```python
filtered['wucha'].min()/60
```

```python
filtered.eval('wucha_min = wucha/60' , inplace=True)
```

```python
import math
filtered['wucha_min_ceil'] = filtered['wucha_min'].apply(lambda x:math.ceil(x) if x>=0 else math.ceil(x-1))
filtered.head()
```

```python
grouped = filtered.groupby('wucha_min_ceil')
result=grouped.describe()['RECORD_ID']
```

```python
result.to_csv('result.csv')
```
