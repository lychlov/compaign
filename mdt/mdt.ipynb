```python
import pandas as pd
```

```python
cells = pd.read_csv("高负荷小区列表.csv")
mdt_cell=pd.read_csv('50.csv')
```

```python
mdt_cell['total_rsp'] = mdt_cell.apply(lambda x:x['AVG_RSRP']*x['GRID_CELL_SAMPLES'],axis=1)
grouped = pd.DataFrame(mdt_cell.groupby('ECI').sum())
```

```python
merged = pd.merge(cells,grouped,right_on='ECI',left_on='小区ECI')
```

```python
merged['SDATE']="2018/4/1"
```

```python
merged['小区弱覆盖率（小于-110dBm）']=merged.apply(lambda x: round(x['POORCOVERAGE110']/x['GRID_CELL_SAMPLES']*10000)/100.0,axis=1)
merged['小区平均RSRP']=merged.apply(lambda x:round(x['total_rsp']/x['GRID_CELL_SAMPLES']*100)/100.0,axis=1)
```
