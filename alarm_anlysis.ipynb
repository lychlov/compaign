```python
standard_alarms=pd.merge(alarms,jizhan,on='alarm_object')
standard_alarms=pd.merge(standard_alarms,xian,on='xian_id')
standard_alarms=pd.merge(standard_alarms,shi,on='shi_id')
standard_alarms.head()
```

```python
standard_alarms['end_time'] =pd.to_datetime(standard_alarms['end_time'])
standard_alarms['start_time'] =pd.to_datetime(standard_alarms['start_time'])
```

```python
zero = pd.to_datetime('2018-08-08 00:00:00')
two_four = pd.to_datetime('2018-08-09 00:00:00')
standard_alarms[(standard_alarms.start_time < two_four)] \
    [standard_alarms.start_time >= zero].groupby('shi_name').describe() \
    ['serial_no'].sort_values(by='count', ascending=False)
```

```python
zero = pd.to_datetime('2018-08-08 00:00:00')
two_four = pd.to_datetime('2018-08-09 00:00:00')
down_alarms=standard_alarms[(standard_alarms.end_time < two_four)] \
    [standard_alarms.alarm_no=='007-103-00-040012']\
    .sort_values(by='start_time')
```

```python
down_alarms.info()
```

```python
down_alarms.tail()
```

```python
new_time_serial = down_alarms[['serial_no', 'start_time', 'shi_name']]
new_time_serial.rename(columns={'start_time': 'time'}, inplace=True)
new_time_serial['type'] = 'start'
new_time_serial.head()
```

```python
new_time_serial_end= down_alarms[['serial_no','end_time','shi_name']]
new_time_serial_end.rename(columns={'end_time': 'time'}, inplace=True)
new_time_serial_end['type']='end'
new_time_serial_end.head()
```

```python
new_serial=new_time_serial.append(new_time_serial_end)
new_serial=new_serial.sort_values(by=['time'])
```

```python
new_serial.to_csv(r'C:\Users\Loyocc\Desktop\new_time_serial.csv')
```

```python
now_down_dict = {'安阳市': {'count': 0, 'status': 0}}
for index, row in new_serial.iterrows():
    if row['type'] == 'start':
        now_down_dict[row['shi_name']]=now_down_dict.get(row['shi_name'],{'count': 0, 'status': 0})
        now_down_dict[row['shi_name']]['count'] =now_down_dict.get(row['shi_name']).get('count') + 1
        current_status = now_down_dict[row['shi_name']].get('status', 0)
        if (now_down_dict[row['shi_name']]['count'] >= 60 and current_status == 0):
            print("%s 发生二级预警 :%s" % (row['shi_name'], row['time']))
            now_down_dict[row['shi_name']]['status'] = 2
        elif (now_down_dict[row['shi_name']]['count'] >= 150 and current_status != 1):
            print("%s 发生一级预警 :%s" % (row['shi_name'], row['time']))
            now_down_dict[row['shi_name']]['status'] = 1
    else:
        now_down_dict[row['shi_name']]['count'] =now_down_dict.get(row['shi_name']).get('count') - 1
        current_status = now_down_dict[row['shi_name']].get('status', 0)
        if (now_down_dict[row['shi_name']]['count'] < 150 and current_status == 1):
            print("%s 解除一级预警 :%s" % (row['shi_name'], row['time']))
            now_down_dict[row['shi_name']]['status'] = 2
        elif (now_down_dict[row['shi_name']]['count'] < 60 and current_status != 0):
            print("%s 解除所有预警 :%s" % (row['shi_name'], row['time']))
            now_down_dict[row['shi_name']]['status'] = 0
```
